```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.formula.api import logit
```

```python
# Question 2

num_list = [3892, 20, -145, 123, 89, -78, -93, 41, -543, -86, 583, -39, 76, 351, -34]

pos_odd = []
pos_even = []
neg_odd = []
neg_even = []

for num in num_list:
  if num > 0 and num % 2 != 0:
    pos_odd.append(num)
  elif num > 0 and num % 2 == 0:
    pos_even.append(num)
  elif num < 0 and num % 2 != 0:
    neg_odd.append(num)
  elif num < 0 and num % 2 == 0:
    neg_even.append(num)
  else:
    print('Error :(')
    continue

print(pos_even,
      pos_odd,
      neg_even,
      neg_odd)
```

```python
# Question 3
diabetes = pd.read_csv('/content/diabetes.csv')

print(diabetes.shape)
diabetes.describe().T
```

```python
# Question 4a

low_preg = []

for diabetic in diabetes['Pregnancies']:
  if diabetic < 3:
    low_preg.append(diabetic)
  else:
    continue

dia_percentage = (len(low_preg)/len(diabetes)) * 100
# print(low_preg)
print(f'{dia_percentage}% of diabetics in the dataset have had less than 3 pregnancies')
```

```python
# Question 4b

bp = []
age = []
bmi_age = []

for BloodPressure, Age, BMI in zip(diabetes['BloodPressure'], diabetes['Age'], diabetes['BMI']):
  bp.append(BloodPressure)
  age.append(Age)
  bmi_age.append(BMI/Age)
```

```python
# Question 5

ins_diag_model = logit('Outcome ~ Insulin', data = diabetes).fit()
print(ins_diag_model.summary())
exp_values = np.exp(ins_diag_model.params)
print(exp_values)

# The exponentiated values of the logistic regression tell us that for every 1 unit increase in insulin level, the odds of diabetes are 1.0023 times more likely.
# There is a base level of 0.44285, meaning at an insulin level of zero, the odds of diabetes are 0.443.
```

```python
# Question 6

inglu_model = logit('Outcome ~ Insulin + Glucose', data = diabetes).fit()
print(inglu_model.summary())
expval = np.exp(inglu_model.params)
print(expval)

# Based on the results of this logistic regression, for every 1 unit increase in insulin level, the odds of diabetes are 0.999 times more likely, assuming that glucose levels are held constant.
# Additionally, for every 1 unit increase in glucose level, the odds of diabetes are 1.040 times more likely, assuming that insulin levels are held constant.
```
